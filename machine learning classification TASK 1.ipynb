```python


# Define the image dimensions for resizing
img_shape = (128, 64, 3)

# Define the list to store flattened image data and category index
data_flat = []
indexs_arr = []

# Define the categories in your dataset
Categories = ['Cars', 'Ice cream cone', 'Cricket ball']

# Define the directory where your images are located
datadir = 'Image Classification dataset'

for i in Categories:
    print(f'Loading category: {i}')
    path = os.path.join(datadir, i)
    
    for img in os.listdir(path):
        img_path = os.path.join(path, img)
        img_array = imread(img_path)
        
        # Resize the image to the specified dimensions
        img_resized = resize(img_array, img_shape)
        
        # Flatten the resized image and store it in data_flat
        data_flat.append(img_resized.flatten())
        
        # Store the index of the category in indexs_arr
        indexs_arr.append(Categories.index(i))
    
    print(f'Loaded category: {i} successfully')

# Convert data_flat and indexs_arr to numpy arrays
data_flat = np.array(data_flat)
target = np.array(indexs_arr)

# Create a DataFrame with flattened image data and target labels
df = pd.DataFrame(data_flat)

# Add the 'Target' column with category labels
df['Target'] = target
```

### ORB extractor

```python
from skimage.feature import orb
orb_vectors = []
labels = []
orb = cv2.ORB_create()
class_labels=['Cars','Ice cream cone','Cricket ball']
# Loop over each image in the directory
for i, class_label in enumerate(class_labels):
    for file in os.listdir(os.path.join(datadir, class_label)):
        if file.endswith(".jpg"):
            # Load the image
            img = cv2.imread(os.path.join(datadir, class_label, file))
            img=cv2.resize(img , (64, 128))
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            # Extract the orb features
            keypoints, descriptors = orb.detectAndCompute(img, None)
            descriptors = np.resize(descriptors,(500,128)).flatten()
            orb_vectors.append(descriptors)
            labels.append(i)
          

  #    Convert list to np.array
orb_vectors = np.array(orb_vectors)

# Convert list feature to nparray
labels = np.array(labels)
df=pd.DataFrame(orb_vectors)

df['Target']=labels
```
